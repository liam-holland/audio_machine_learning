## 3 - Logging Data to wandb

We can log things to wandb, like, the training loss, validation loss, accuracy, or pretty much anything! Let's try.

```python
wandb.log({"epoch": 0, "train-loss": 41.2})
```

This should create some graphs on wandb, showing the 'train-loss' against 'step'. Lets log some more:

```python
for n in range(1,20): # Simulating training a neural network!
    wandb.log({"epoch": n, "train-loss": 41.2-n-(2*np.random.randn())}) # Simulate our loss decreasing with epoch.
```

It might take a 10-20 seconds to update, but now you should see the updated 'train-loss' graph.

During your model training you should also log validation and test loss, in a similar fashion. 

## 4 - Creating Multiple Runs in wandb

Great, so you can log stats about your model run, what next? Let's try a new run, we'll pretend we made some great modification to our method, and now the loss goes down faster!

We will start a new run, with 'reinit=True', this will end the previous run and start a new one.

```python
wandb.init(project=project_name, reinit=True) # Use the same project name, so it is pushed to the same project on wandb.
```

Now look at wandb, and you can see the new run is visible. Lets log some values to it.

```python
for n in range(0,20):
    wandb.log({"epoch": n, "train-loss": 41.2-n*1.5-(2*np.random.randn())}) # Simulate our loss decreasing with epoch.
```

You should now be able to see 'train-loss' curves for both runs, on the same graph. You can keep doing this with as many runs as you like. There are two great benefits to this:
- Firstly it gives you a simple visual to see if your training is progressing correctly. 
- Secondly, it allows you compare different runs easily!

## 5 - Logging Audio to wandb

It is also possible to log audio to wandb. 

For example, maybe you want to be able to easily listen to incorrectly classified examples from your test/validation set.

Here is an example of pushing audio to wandb:

```python
# Create some audio!
length_s = 2
freq = 440
fs = 44100
time = np.arange(0, length_s, 1/fs)
wave = np.sin(time * 2 * np.pi * freq)

# Log the audio!
wandb.log({'incorrectly_classified':  wandb.Audio(wave, sample_rate=44100, caption='Correct Class = ')})
```

This should shortly appear under 'media', on wandb. 

## 6 - Logging Hyperparameters

When initiating a project, you want to be able to see what hyperparameters were used on that run. I suggest doing this by keeping all run parameters in a dictionary. When initiating your run, you can add this config dictionary. You can then filter your runs on wandb based on these config settings. Here is an example.

```python
run_config = {'num_layer': 5, 'hidden_size': 64} #Here are the hyperparameters for this run
```

```python
wandb.init(project=project_name, reinit=True , config=run_config) 
```

If you go to Runs, there is a filter option, which allows you to filter runs by various configs. This is very useful if you have lots of runs.

## 7 - Further Work

Hopefully this gave you some idea of how using a service like wandb can be helpful for your projects (or any other ML projects you might be working on!). Here are some things you might consider logging when training your models:

Basic Training Metrics:
- Loss values (training/validation/test loss)

Hyperparameter & Experiment Tracking:
- Batch size
- Data augmentations applied
